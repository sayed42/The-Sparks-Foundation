How to predict the percentage of a student based on the no of study hours and what will be the predicted score if a student studies for 9.25 hours/day

**Importing required Libraries**


```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

```

**Importing Data Set**

```python
df=pd.read_csv("/content/student_scores - student_scores.csv")
```

```python
sns.regplot(df.Hours,df.Scores)
```

```python
x=df[['Hours']]
y=df['Scores']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size = 0.2,random_state=10)
```

```python
m = LinearRegression()
m.fit(x_train,y_train)
print('Model Fitted')
```

```python
y_pred = m.predict(x_test)
y_pred
```

```python
output = pd.DataFrame({'Actual Values':y_test,'Predicted Values':y_pred,'Error':abs(y_test-y_pred)})
output
```

```python
m.score(x_test,y_test)
```

**Solving the given Problem**

```python
m.predict([[9.25]])[0]
```
